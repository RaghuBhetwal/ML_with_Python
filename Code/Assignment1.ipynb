```python
# merging all seperate dataframe into single dataframe as df
con1 = pd.merge(result_df, races_df, on ='raceId')
con2 = pd.merge(con1, drivers_df, on = 'driverId')
con3 = pd.merge(con2, driver_standings_df, on = 'driverId')
con4 = pd.merge(con3, constructor_df, on ='constructorId')
df = pd.merge(con4, stats_df, on ='statusId')
pd.get_option("display.max_columns",None)
df.head()
```

```python
# droping unwanted columns 

df = df.drop(['url_y','url_x','position_x','fastestLapTime','positionText_x','driverRef',
              'constructorRef','nationality_y','positionText_y','raceId_y','points_y'],1)
```

```python
# changing of old column name to meaningful name 

col_name = {'number_x':'number','milliseconds':'timetaken_in_millisec','fastestLapSpeed':'max_speed',
 'name_x':'grand_prix','number_y':'driver_num','code':'driver_code','nationality_x':'nationality','name_y':'company',
 'raceId_x':'racerId','points_x':'points','position_y':'position'}
```

```python
df.rename(columns=col_name,inplace=True)
df.head()
```

```python
# Adding two column (forename,surname) to create a single column called driver_name

df['driver_name'] = df['forename']+' '+df['surname']
```

```python
df = df.drop(['forename','surname'],1)
```

```python
data = df[["driver_name", "company", "position", "status", "nationality", "wins", "rank","points","timetaken_in_millisec"]]
data.head()
```

## Data Visualization

```python
numerical_cols = ['wins', 'rank', 'points', 'timetaken_in_millisec']

corr_matrix = data[numerical_cols].corr()

ax = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

ax.set_title("Correlation Matrix of Numerical Columns", fontsize=14)

plt.show()
```

```python

wins_by_company = data.groupby('company')['wins'].sum().reset_index()

top_10_companies = wins_by_company.sort_values('wins', ascending=False).head(10)

ax = sns.barplot(x='wins', y='company', data=top_10_companies)

plt.title('Top 10 Companies by Total Wins')

ax.set_xlabel("Total Wins", fontsize=12)

ax.set_ylabel("Company", fontsize=12)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

wins_by_nationality = data.groupby('nationality')['wins'].sum().reset_index()

wins_by_nationality = wins_by_nationality.sort_values('wins', ascending=False)

top_10_nationalities = wins_by_nationality[:10]

ax = sns.barplot(x='nationality', y='wins', data=top_10_nationalities)

ax.set_xlabel("Nationality", fontsize=12)

ax.set_ylabel("Number of Wins", fontsize=12)

ax.set_title("Top 10 Nationalities by Wins", fontsize=14)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.show()
```

```python

cols = ['driver_name', 'company', 'nationality', 'wins']

grouped = data[cols].groupby(['company', 'nationality']).sum().reset_index()

pivot = grouped.pivot(index='nationality', columns='company', values='wins')

# top 10 companies by total number of wins
top_10_companies = pivot.sum().sort_values(ascending=False)[:10].index.tolist()
pivot_top_10 = pivot[top_10_companies]

ax = pivot_top_10.plot(kind='bar', stacked=True, figsize=(12, 8))

ax.set_title('Number of Wins by Company and Driver Nationality (Top 10 Companies)', fontsize=14)
ax.set_xlabel('Driver Nationality', fontsize=12)
ax.set_ylabel('Number of Wins', fontsize=12)

# Show the plot
plt.show()
```
